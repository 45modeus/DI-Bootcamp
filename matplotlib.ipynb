```python
import pandas as pd

df = pd.DataFrame({
    'Month':['Jan', 'Feb', 'Mar', 'Apr', 'May', 
            'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec'],
    'Temperature':[5.8, 6.1, 9.2, 12.8, 16.8, 19.9, 22.2, 22.4, 19.2, 14.6, 9.8, 6.5]
})

df.plot(kind='scatter',x='Month',y='Temperature',color='blue')
plt.show()
```
