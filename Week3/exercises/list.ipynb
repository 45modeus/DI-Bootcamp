```python
basket = ["Banana", "Apple", "Oranges", "Blueberries"]
print("Original basket:", basket)
```

```python
basket.remove("Banana")
basket.remove("Blueberries")
print("After removing Banana and Blueberries:", basket)
```

```python
basket.append("Kiwi")
print("After adding Kiwi:", basket)
```

```python
apple_count = basket.count("Apple")
print("Number of times 'Apple' appear:", apple_count)
```
