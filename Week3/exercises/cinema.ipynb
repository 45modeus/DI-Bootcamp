```python
ages = []

num_people = int(input("How many people are buying tickets? "))

for i in range(num_people):
    age = int(input(f"Enter the age of person {i + 1}: "))
    ages.append(age)

total_cost = 0

for age in ages:
    if age < 3:
        price = 0
    elif 3 <= age <= 12:
        price = 10
    else:
        price = 15
    total_cost += price

print(f"\nTotal ticket cost: ${total_cost}")
```
